```python
''' 내가 푼 1번째 '''
result = []
a = 0
for _ in range(7):
    num = int(input())
    if num % 2 != 0: result.append(num)
    else: a += 1

print(-1) if a == 7 else print(sum(result), '\n', min(result), sep='')
```

```python
''' 내가 품 2번째 '''
lst = [int(input()) for _ in range(7)]
result = [i for i in lst if i % 2 != 0]

if len(result) == 0: print("-1")
else: print(sum(result), "\n", min(result), sep='')
```

```python
lst = [int(input()) for _ in range(7)]
result = [i for i in lst if i % 2 != 0]

print(f"{sum(result)}\n{min(result)}" if result else -1)  # 중요
```

```python
''' 다른사람 - 숏코딩 '''

a = [x for _ in range(7) if (x := int(input())) % 2]

print(f"{sum(a)}\n{min(a)}" if a else -1)    #(중요) if를 print문 안에도 넣을 수 있구나
```
