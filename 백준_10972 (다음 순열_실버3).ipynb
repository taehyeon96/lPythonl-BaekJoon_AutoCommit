```python
from itertools import permutations

n = int(input())
arr = list(map(int, input().split()))
ans = [i for i in arr]  # 정렬 후 같이 저장되어버려서 이렇게함

memory = []

arr.sort()

for i in permutations(arr):
    memory.append(i)

leng = len(memory)

for i in range(leng):
    if memory[i] == memory[-1]:
        print(-1)
    elif ans == list(memory[i]):
        print(" ".join(map(str,memory[i+1])))
        break
            
```

```python
from itertools import permutations

n = int(input())
arr = list(map(int, input().split()))
ans = [i for i in arr]  # 정렬 후 같이 저장되어버려서 이렇게함
ans_reverse = [i for i in arr]  # 정렬 후 같이 저장되어버려서 이렇게함

arr.sort()  # 테케 2번같은 입력을 permutations하면 결과가 안나와서 정렬해줌
ans_reverse.sort(reverse = True)

flag = False

for i in permutations(arr):
    if flag == True:
        print(" ".join(map(str,i)))
        break
        
    if list(i) == ans_reverse:
        print(-1)
    elif list(i) == ans:
        flag = True
```

```python
# https://ji-gwang.tistory.com/263?category=1044736

n = int(input())
arr = [i+1 for i in range(n)]
arr_reversed = sorted(arr, reverse=True)

def permutation(arr):
    print(" ".join(map(str,arr)))  # 일단 출력
    
    # 마지막 순열인 경우 - 종료
    if arr == arr_reversed:
        exit()
        
    for i in range(n-1, 0, -1):   # 끝에서부터 앞 값이랑 비교하기 위해 idx=1까지 반복
        if arr[i-1] < arr[i]:
            for j in range(n-1, 0, -1):  # 여기도 다름
                if arr[i-1] < arr[j]:
                    arr[i-1], arr[j] = arr[j], arr[i-1]
                    arr = arr[:i] + sorted(arr[i:])
                    # 여기가 달라진다!!! - 재귀호출
                    permutation(arr)
                
permutation(arr)
```

```python
# https://kjhoon0330.tistory.com/15
def permutation(arr, n):
    result = []
    if n == 0:
        return [[]]
    
    for i in range(len(arr)):
        element = arr[i]
        # i를 제외한 나머지로 재귀호출
        for rest in permutation(arr[:i] + arr[i+1:], n - 1):
            result.append([element] + rest)
    return result


print("\n".join(map(str,permutation([1, 2, 3], 3))))
```
