```python
''' 내가 풀다가 못풀었음 (같은 알파벳 두 개 이상 나올 경우 error) '''
from string import ascii_lowercase

s = input()
arr = [-1] * 21

for i in range(len(s)):
    arr[ascii_lowercase.find(s[i])] = i
    
print(' '.join(map(str, arr)))
```

```python
''' 블로그  https://ooyoung.tistory.com/68 '''
s = input()
alpha = list(range(97,123))  #  (중요) 아스키코드 숫자 범위

for i in alpha :
    print(s.find(chr(i)), end=' ')  # i가 s에서 몇 번째 인덱스인지 순차적으로 출력
```
