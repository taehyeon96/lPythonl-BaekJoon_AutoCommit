```python
''' 내가 푼 (for_ in range(t)이거 뭔 내용인지 몰라서 블로그 참고함) '''

t = int(input())

for _ in range(t):                   # 테스트케이스를 이만큼 주겠다는 뜻이었음
    a = list(bin(int(input())))[2:]
    a.reverse()
    
    for i in range(len(a)):
        if a[i] == '1': 
            print(i, end=' ')
    
```

```python
''' 블로그 풀이 -> 근데 결국 인덱스 구하는건 똑같음, 내장함수 사용하냐 마냐 차이 '''
T = int(input())

for _ in range(T):
    n = bin(int(input()))[2:]

    for idx, val in enumerate(n[::-1]):   # enumerate() 하면 인덱스와 값으로 나눠줌(중요)
        if val == '1':
            print(idx, end=' ')
```
