```python
''' 백준 2164 - 내가 품 => 리스트로 풀 경우 시간초과 판정 발생할 수 있다 '''

from collections import deque

N = int(input())
deque = deque([i for i in range(1, N+1)])  # 데크 생성

while len(deque) > 1:
    deque.popleft()
    deque.rotate(-1)      # 음수일 경우 왼쪽으로 숫자만큼 이동
    
print(deque[0])
```

```python
from collections import deque

n = int(input())
que = deque([i for i in range(1,n+1)])


while que:
    if len(que) > 1:
        que.popleft()
    else:
        print(que[0])
        break        
    que.append(que.popleft())
```
