We want to win the first prize

```python
''' 내가 푼 '''
arr = [list(input().split()) for _ in range(int(input()))]

for i in range(len(arr)):
    for j in range(len(arr[i])):
        print(str(arr[i][j])[::-1], end=" ")
    print()
```

```python
''' 다른 사람 정답 - 입력 받을 때 {문자열 거꾸로} 후 {전체 거꾸로} '''
for i in range(int(input())):
    word = input()[::-1].split()   # ['yadot', 'yppah', 'ma', 'I'], ['ezirp', 'tsrif', 'eht', 'niw', 'ot', 'tnaw', 'eW']
    word.reverse()                 # ['I', 'ma', 'yppah', 'yadot'] , ['eW', 'tnaw', 'ot', 'niw', 'eht', 'tsrif', 'ezirp']
    
    print( ' '.join(word))         # I ma yppah yadot
```
