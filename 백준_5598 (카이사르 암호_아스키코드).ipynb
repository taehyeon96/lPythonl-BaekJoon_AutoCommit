```python
''' 내가 푼 '''

for i in input():
    if i == 'A': print('X', end='')
    elif i == 'B': print('Y', end='')
    elif i == 'C': print('Z', end='')
    else: print(chr(ord(i)-3), end='')
```

```python
''' 다른 사람이 푼 - 숏코딩'''
for i in input():print(chr((ord(i)-42)%26+65),end='')
```
