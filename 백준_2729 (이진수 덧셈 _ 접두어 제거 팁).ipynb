

- 링크 : https://www.acmicpc.net/problem/2729

[예제 입력 1]
3
1001101 10010
1001001 11001
1000111 1010110

* 예를들어 첫 번째 줄의 경우, 1001101 와 10010 을 더해주는 것임

```python
""" 내가 푼 답 """
result = []
for _ in range(int(input())):
    a, b = map(str, input().split())
    temp = str(bin(int(a,2)+int(b,2)))
    result.append(temp[2:])

print("\n".join(map(str,result)))
```

```python
""" 접두어 제거 - 내가 푼 답 """
result = []
for _ in range(int(input())):
    a, b = map(str, input().split())
    result.append(bin(int(a,2)+int(b,2))[2:])

print("\n".join(map(str,result)))
```
