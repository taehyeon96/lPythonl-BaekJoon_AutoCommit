```python
''' 내가 품 - 블로그 참고하여 => 시간초과 판정 '''
data = []
for _ in range(int(input())):
    data.append(list(map(int,input().split())))

data.sort(key=lambda x:(x[0], x[1]))   # 핵심 (중요)

for i in data:
    print(i[0], i[1])   # => 이거 때문에 시간초과 판정 발생
```

## (중요) 람다 식 참고
 * https://covenant.tistory.com/142
 * x[0] (=x좌표)를 기준으로 정렬하고, x[0]이 같다면, x[1] (=y좌표)를 기준으로 오름차순 정렬
     * 내림차순 정렬을 하고싶다면, -x[0]

```python
''' 블로그 정답 => 시간초과 판정 아님 '''
arr = []
for _ in range(int(input())):
    arr.append(list(map(int, input().split())))

arr.sort(key=lambda x:(x[0], x[1]))
 
for e in arr:
    print(str(e[0]) + " " + str(e[1]))
```
