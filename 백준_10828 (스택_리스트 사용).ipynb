```python
''' 내가푼 - 시간초과 '''
# 시간초과 나온다면 1. input을 sys로 바꾸기  2. deque로 바꾸기
result = []
for _ in range(int(input())):
    cmd = input()
    
    if cmd[:3] == "pus": result.append(cmd.split()[1])
    elif cmd == "pop" and result:
        print(result[-1])
        result = result[:-1]
    elif cmd == "pop" and not result:
        print(-1)
    elif cmd == "size": print(len(result))
    elif cmd == "empty" and result: print(0)
    elif cmd == "empty" and not result: print(1)
    elif cmd == "top" and not result: print(-1)
    elif cmd == "top" and result: print(result[-1])
    
```

```python
''' 내가 푼 - 시간초과 해결 '''
# 시간초과 나왔으니 1. input을 sys로 바꾸기
import sys

result = []
for _ in range(int(sys.stdin.readline())):
    cmd = sys.stdin.readline()
    
    if cmd[:3] == "pus": result.append(cmd.split()[1])
    elif cmd == "pop" and result:
        print(result.pop())
    elif cmd == "pop" and not result:
        print(-1)
    elif cmd == "size": print(len(result))
    elif cmd == "empty" and result: print(0)
    elif cmd == "empty" and not result: print(1)
    elif cmd == "top" and not result: print(-1)
    elif cmd == "top" and result: print(result[-1])
    
```

```python
''' 블로그 '''
import sys

# 정수 X를 스택에 넣는 연산이다.
def push(x):
    stack.append(x)

# 스택에서 가장 위에 있는 정수를 빼고, 그 수를 출력한다. 
# 만약 스택에 들어있는 정수가 없는 경우에는 -1을 출력한다.
def pop():
    if(not stack):
        return -1
    else:
        return stack.pop()

# 스택에 들어있는 정수의 개수를 출력한다.
def size():
    return len(stack)

# 스택이 비어있으면 1, 아니면 0을 출력한다.
def empty():
    return 0 if stack else 1

# 스택의 가장 위에 있는 정수를 출력한다. 
# 만약 스택에 들어있는 정수가 없는 경우에는 -1을 출력한다.
def top():
    return stack[-1] if stack else -1

N = int(sys.stdin.readline().rstrip())
stack = []

for _ in range(N):
    input_split = sys.stdin.readline().rstrip().split()

    order = input_split[0]

    if order == "push":
        push(input_split[1])
    elif order == "pop":
        print(pop())
    elif order == "size":
        print(size())
    elif order == "empty":
        print(empty())
    elif order == "top":
        print(top())
```
