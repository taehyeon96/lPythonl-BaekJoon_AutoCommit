

* 링크 : https://www.acmicpc.net/problem/2490

0 1 0 1
1 1 1 0
0 0 1 1

```python
''' 내가 푼 '''
result = []

for _ in range(4):
    yut = list(map(int,input().split()))    
    result.append({1:'A', 2:'B', 3:'C', 4:'D', 0:'E'}.get(yut.count(0), "default")) # (중요) 0:E는 "모"의 경우임

print("\n".join(result))
```

* if-else 쓰기 싫어서 switch-case 참고한 링크 : https://thrillfighter.tistory.com/602

```python
''' 블로그 답 - 이 방법이 싫어서 위와 같이 짠 것임 '''
def check(arr):
    num_of_zero = arr.count(0)
    if num_of_zero == 0:
        print("E")
    elif:
        pass

for i in range(3):
    arr = list(map(int, input().split()))
    check(arr)
```
