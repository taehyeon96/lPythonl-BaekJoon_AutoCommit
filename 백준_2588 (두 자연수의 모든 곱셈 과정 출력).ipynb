```python
''' 백준 2588 내가 품 '''
a = int(input())
b = list(map(int,input()))

j = 0
result = 0

for i in range(len(b)-1,-1, -1):    # 이것 쓰지 말고 b.reverse()를 활용하자
    print(a * b[i])
    result = result + ((a * b[i]) * pow(10, j))
    j = j + 1

print(result)
```

```python
''' 백준 2588 reverse()활용 '''
a = int(input())
b = list(map(int,input()))

b.reverse()

result = 0

for i in range(len(b)):
    print(a * b[i])
    result = result + ((a * b[i]) * pow(10, i))

print(result)
```
