3의 이웃 = 2, 4, 2x3 = [-1, 1, x2] = dx 가 된다

0같은 경우엔 왼쪽이 없는데, 필요없고 if < or >= or < or >= 하면 신경 꺼도 된다

이동할 때마다 cnt += 1

그렇다면 dfs일까 bfs일까
 -> 이웃이 3가지 케이스이니까 bfs인거같고, 이동할 때마다 cnt += 1
 -> 뭔가 안되면 이동 횟수 기록하는 [0,0,0,0 ... ]를 만들고 min()하는 것도 방법임

```python
''' 내가 푼 - 1차 - 답이아님(백업) '''
from collections import deque
x, k = map(int, input().split())
graph = [i for i in range(k+1)]

que = deque([x])  # 시작점은 수빈이의 위치 (= x인데 0부터 시작)

dx = [-1, 1, 2]

cnt = 0

# BFS는 재귀가 없다 항상 주의해라
while que:
    x = que.popleft()
    for i in dx:
        nx = x + i if i < 2 else x * 2
        
        cnt += 1
        
        if nx < 0 or k < nx:
            continue
            
        if nx == k:
            que.clear()  # = break
            break
        else:
            que.append(nx)
        
print(cnt) 
```
