* 링크 : https://www.acmicpc.net/problem/11005

입력)
60466175 36

```python
''' 내가 짠 (아래와 동일 문제점 발생) '''
n, b = map(int, input().split())

result = []
def change(n, b):
    while True:
        result.append(n % b)
        if n // b < b:
            result.append(n//b)
            return result
        else:
            return change(n//b, b)
        

print(change(n,b))
```

```python
''' 블로그 코드 (내것과 동일 문제점 발생) '''
n, b = map(int, input().split())

def base10(n,b):
    if n==0: return [0]
    
    result = []
    while n>0:
        n, m = divmod(n, b)   # (중요) n=몫,  m=나머지
        result.insert(0, m)   # 우리가 필요한 숫자는 나머지(m)
    
    return result

print(base10(n,b))
```

## (중요) 위 코드는 10진수->0~9진수만 처리 가능
##             즉, 10진수 이상은 아래와 같이 처리해야 함



```python
''' (중요) 블로그 코드 이어서'''
def encode(digits, digit_map):    # digits=change()의 리턴값, digit_map=맵핑할 문자열
    encoding = ''    
    for d in digits:
        encoding += digit_map[d]
    return encoding

print(encode([15,15], '0123456789ABCDEF'))
```

```python
''' 최종 블로그 코드 '''
n, b = map(int, input().split())

def total_base10(n,b):

    digit_map = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'   # 전부 순서대로
    
    if b < 2 or b > 36:  # 36이상은 안됨!
        raise ValueError("invalid base: 2<=b<=36")
    
    digits = base10(n, b)
    encoding = encode(digits, digit_map)
    
    return encoding

print(total_base10(n,b))
```

```python
''' 풀어써보자 '''
n, b = map(int, input().split())

def total(n,b):    
    result = []
    
    while n > 0:
        n, m = divmod(n, b)   # (중요) n=몫,  m=나머지
        result.insert(0, m)   # 우리가 필요한 숫자는 나머지(m)
    
    digit_map = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'   # 전부 순서대로
    
    encoding = ''    
    for d in result:
        encoding += digit_map[d]
        
    return encoding

print(total(n,b))
```
