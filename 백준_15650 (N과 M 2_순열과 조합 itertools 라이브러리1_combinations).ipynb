```python
''' 백준 15650 - 내가 푼(블로그 참고) '''
from itertools import combinations    # iter tools의 조합(콤비네이션s) (<--> 순열 permutations)

n, m = map(int, input().split())

# 리스트를 따로 만들지 않고 한줄에 해줬으며, (중요) 이때 각 원소를 str()로 변환해줘야 함
result = list(combinations([str(i) for i in range(1, n+1)], m)) 

for i in result:
    print(" ".join(i))

#                첫 번째 인자 = 리스트, 두 번째 인자 = nCm일 경우, m
```

```python
''' 참고) str()안해줄 경우 '''
n, m = map(int, input().split())

# str()안했을 경우
result = list(combinations([i for i in range(1, n+1)], m)) 

for i in result:
    print(" ".join(i))
```

```python
''' 블로그 정답 '''
from itertools import combinations  
 
N, M = map(int, input().split())  
arr = [str(i+1) for i in range(N)]   # 각 원소 str()로 만들어줘야함
 
for e in list(combinations(arr, M)):  
    print(" ".join(e))
```
